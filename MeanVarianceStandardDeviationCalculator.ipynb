```python
import numpy as np
```

```python
def calculate(list):
    if len(list) == 9:
        list = np.array(list).reshape(3,3)
        print('mean :', np.mean(list, axis=0), np.mean(list, axis=1), np.mean(list))
        print('variance :', np.var(list, axis=0), np.var(list, axis=1), np.var(list))
        print('standard deviation :', np.std(list, axis=0), np.std(list, axis=1), np.std(list))
        print('max :', np.max(list, axis=0), np.max(list, axis=1), np.max(list))
        print('min :', np.min(list, axis=0), np.min(list, axis=1), np.min(list))
        print('sum :', np.sum(list, axis=0), np.sum(list, axis=1), np.sum(list))

    else:
        raise ValueError('List must contain nine numbers.')


    return calculate
```

```python
calculate([0,1,2,3,4,5,6,7,8])
```

```python
calculate([1,5,9,7,8])
```
